```python
#No need  of Id column
data.drop("id", axis=1,inplace=True)
data.head()
```

```python
#looking into diagnosis column
data['diagnosis'].unique()
```

```python
data['diagnosis'].nunique()
```

```python
data.loc[data['diagnosis']=='M','diagnosis']=0
data.loc[data['diagnosis']=='B','diagnosis']=1
data.head()
```

```python
data['diagnosis'].unique()
```

```python
#Feature selection
X=data.drop("diagnosis", axis=1)
y=data['diagnosis']
```

```python
X.columns
```

```python
y=np.array(y, dtype=int)
```

```python
#Split the dataset for training & testing 
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
#Scaling 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
model=RandomForestClassifier(n_estimators=100,criterion="entropy",random_state=42)
result=model.fit(X_train,y_train)
```

```python
data.head()
```

```python
y.dtype
```

```python
X_train.dtype
```

```python
pred=result.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
acc=accuracy_score(y_test,pred)
print("Accuracy Score: ", acc)
```

```python
matrix=confusion_matrix(y_test,pred)
matrix_df=pd.DataFrame(matrix,index=['Actual 0','Actual 1'],columns=['Predicted 0','Predicted 1'])
sns.heatmap(matrix_df,annot=True)
plt.show()
```

```python
print(classification_report(y_test, pred))
```
